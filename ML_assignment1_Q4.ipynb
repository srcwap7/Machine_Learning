```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv",delimiter=";")
```

```python
print(data)
```

```python
data.loc[data["quality"]>6]
```

```python
data.loc[data["quality"] <= 6, "quality"] = 0
```

```python
data.loc[data["quality"] > 6,"quality"] = 1
```

```python
print(data)
```

```python
data.loc[data["quality"]==1]
```

```python
scaler = MinMaxScaler()
```

```python
data_scaled = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
```

```python
print(data_scaled)
```

```python
distinct_values = data_scaled['quality'].unique()
print(distinct_values)
```
